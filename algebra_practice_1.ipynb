```python
import numpy as np
import matplotlib.pyplot as plt


t = np.linspace(0,40,1000)

r_t = 2.5*t
c_t = 3*(t-5)

fig,ax = plt.subplots()
plt.title("Theif catching")
plt.xlabel('Time(minutes)')
plt.ylabel('Distance')
ax.set_xlim([0,50])
ax.set_ylim([0,100])
ax.plot(t,r_t,c='green')
ax.plot(t,c_t,c='red')
ax.vlines(30,ymin=0,ymax=75,color='blue',linestyles='dotted')
ax.hlines(75,xmin=0,xmax=30,color='blue',linestyles='dotted')


```

```python
import numpy as np
import matplotlib.pyplot as plt

d = np.linspace(0,100,1000)
m1 = (d-30)
m2 = 4*d

fig,ax = plt.subplots()
plt.title('total solar energy')
plt.xlabel('kilo watts')
plt.ylabel('days')
ax.set_xlim([0,1000])
ax.set_ylim([0,200])
ax.plot(d,m1,c='green')
ax.plot(d,m2,c='blue')
```
